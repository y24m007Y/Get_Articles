```python
#trainデータの作成

def create_train_data(cate):
    train_df = None
    target_file = [file.name for file in dirpath.iterdir() if cate in file.name]
    for filename in target_file:
        filepath = dirpath.joinpath(filename)
        df = pd.read_csv(filepath, encoding='utf-8')
        print(df.shape[0])
        drop_index = []
        for i in range(len(df)):
            temp_tag = df.tags[i]
            temp_tag = re.sub("[\[\]\' ]", "", temp_tag)
            temp_tag = temp_tag.split(",")
            count = 0
            for tag in temp_tag:
                for key, value in sample_tag.items():
                    if tag in value:
                        if key == cate:
                            break
                        else:
                            drop_index.append(i)
                            break
        print(drop_index)
        df = df.drop(index=drop_index)
        if train_df is None:
            train_df = df
        else:
            train_df = pd.concat((train_df, df), axis=0)
    train_df = train_df.reset_index()
    return train_df

    
```

```python
cate = "データ分析"
sample = create_train_data(cate)

print(sample.shape)
sample.head(10)
```

```python
outputpath = Path("data/Qiita/Train")

if not os.path.exists(outputpath):
    os.makedirs(outputpath)

sample.to_csv(outputpath.joinpath(f"Qiita2022-2023{cate}.csv"), header=True, index=False)
```
